```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
dataset=pd.read_csv("C:/Users/khush/Downloads/Machine Learning Full/KNN/KNN_Dataset.csv")
```

```python
len(dataset)
```

```python
dataset.head()
```

```python
#replacing zeroes from the below columns
zero_replace=['Glucose','BloodPressure','SkinThickness','BMI','Insulin']
for column in zero_replace:
    dataset[column]=dataset[column].replace(0,np.nan)
    mean=int(dataset[column].mean(skipna=True))
    dataset[column]=dataset[column].replace(np.nan,mean)
```

```python
dataset.head()
```

```python
#splitting dataset into train and test set
X=dataset.iloc[:,0:8]
y=dataset.iloc[:,8]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#feature scaling the input independent variables
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
#value of K
import math
math.sqrt(len(y_test))
```

```python
 #define the model i.e. initialise the KNN Classifier
classifier=KNeighborsClassifier(n_neighbors=11,p=2,metric='euclidean')
classifier.fit(X_train,y_train)
```

```python
#predict the test set results
y_pred=classifier.predict(X_test)
y_pred
```

```python
#evaluating the model using con_mat
con_mat=confusion_matrix(y_test,y_pred)
print(con_mat)
print(f1_score(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```
