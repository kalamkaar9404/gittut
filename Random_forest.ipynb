```python
#lib with iris dataset
from sklearn.datasets import load_iris
#loading scikits random forest classifier lib
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
#setting random seed
np.random.seed(0)
```

```python
#creating obj for iris data
iris=load_iris()
#creating a dataframe with 4 feature variable
df=pd.DataFrame(iris.data,columns=iris.feature_names)
#viewing
df.head()
```

```python
#adding a new column for name of specie
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
df.head()
```

```python
df['is_train']=np.random.uniform(0,1,len(df))<=0.75
df.head()
```

```python
#creating df for test and train rows
train,test=df[df['is_train']==True],df[df['is_train']==False]

len(test)
```

```python
len(train)
```

```python
#create a list of feature columns names
features=df.columns[:4]
features
```

```python
#converting species names into digits
y=pd.factorize(train['species'])[0]
y
```

```python
#creating rforest classifier
clf=RandomForestClassifier(n_jobs=2,random_state=0)
#training the classifier
clf.fit(train[features],y)
```

```python
#applying trained classifier to test
clf.predict(test[features])
```

```python
#viewing predicted probailities of first 10 obs
clf.predict_proba(test[features])[0:10]
```

```python
# mapping names
preds=iris.target_names[clf.predict(test[features])]
preds[0:5]
```
