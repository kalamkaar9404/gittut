```python
#importing lib
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
#loading the dataset
bal_data=pd.read_csv("C:/Users/khush/Downloads/Machine Learning Full/Decision Tree/Decision_Tree_ Dataset.csv",sep=',',header= 0)
```

```python
bal_data=bal_data.drop('sum',axis=1)
```

```python
bal_data.head()
```

```python
#separating the target var
X=bal_data.iloc[:,0:4].values
Y=bal_data.iloc[:,4].values
#splitting into test and train
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=100)
#perform training with entropy function
class_entropy=DecisionTreeClassifier(criterion="entropy",random_state=100,max_depth=3,min_samples_leaf=5)
class_entropy.fit(X_train,y_train)
```

```python
#function to make predictions
y_pred_en=class_entropy.predict(X_test)
y_pred_en
```

```python
#checking accuracy
accuracy_score(y_test,y_pred_en)*100
```
